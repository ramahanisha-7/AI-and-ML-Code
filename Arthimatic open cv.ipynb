```python
#importing libraries
import numpy as np
import cv2
import argparse
import matplotlib.pyplot as plt
```

```python
def plt_imshow(title,image):
  image=cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
  plt.show()
  plt.grid(False)
  plt.axis("On")
  plt.title(title)
  plt.imshow(image)
```

```python
args = {
    "image":"canoe_water_nature_221611.jpg"
}
```

# **Lighten and darken**

```python
image=cv2.imread(args["image"])
plt_imshow("original",image)
```

```python
added=cv2.add(np.uint8([200]),np.uint8([100]))
subtracted=cv2.subtract(np.uint8([20]),np.uint8([100]))
print(added)
print(subtracted)
```

```python
added=np.uint8([20])+np.uint8([100])
subtracted=np.uint8([10])-np.uint8([100])
print(added)
print(subtracted)
```

```python
M=np.ones(image.shape,dtype="uint8")*100
added=cv2.add(image,M)
plt_imshow("Lighter",added)
```

```python
M=np.ones(image.shape,dtype="uint8")*100
subtract=cv2.subtract(image,M)
plt_imshow("Darker",subtract)
```

# **BITWISE**

Notes:
1. Given pixel is turned "off" when its value is 0
2. Turned "On" if pixel is greater than 0

```python

rectangle = np.zeros((300, 300), dtype="uint8")
cv2.rectangle(rectangle, (25, 25), (275, 275), 255, -1)
plt_imshow("Rectangle", rectangle)

circle = np.zeros((300, 300), dtype = "uint8")
cv2.circle(circle, (150, 150), 150, 255, -1)
plt_imshow("Circle", circle)
```

```python
OR=cv2.bitwise_or(rectangle,circle)
plt_imshow("OR",OR)
```

```python
AND=cv2.bitwise_and(rectangle,circle)
plt_imshow("AND",AND)
```

```python
NOT=cv2.bitwise_not(rectangle)
plt_imshow("NOT",NOT)
```

```python
XOR=cv2.bitwise_xor(rectangle,circle)
plt_imshow("XOR",XOR)
```

2. values, 0 and 255 -- pixels with a value of 0 (background) are
