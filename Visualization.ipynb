# **Titanic dataset** (titanic.csv)

```python
#Loading the dataset
!wget "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
```

```python
#importing libraries
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import cv2
import imutils
import argparse
%matplotlib inline
```

```python
#Reading the csv file and printing the file
dataset=pd.read_csv("titanic.csv")
```

```python
#printing the dataset
print(dataset)
```

```python
#reading first five rows of the dataset to visulaize the data
dataset.head()
```

```python
#Here we get the numerical values of the dataset
dataset.describe()
```

```python
#visualising the survivours in the regions
#C — Cherbourg, Q — Queenstown, S — Southampton
dataset.groupby('Embarked').Survived.mean().plot(kind='pie')
```

```python
#survivors on titanic males and femals
dataset.groupby('Sex').Survived.mean().plot(kind='bar')
```

```python
#Survivours according to the Pclass (classes)
dataset.groupby('Pclass').Survived.mean().plot(kind='line')
```

```python
#counting the nan values
dataset.isnull().sum()
```

# **Image file** (jpg)

```python
!wget "https://5.imimg.com/data5/QU/GI/EL/ANDROID-12245496/product-jpeg-500x500.jpg"
```

```python
def plt_imshow(title, image):
	# convert the image frame BGR to RGB color space and display it
	image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
	plt.imshow(image)
	plt.title(title)
	plt.grid(False)
	plt.show()
```

```python
args={
    "image":"product-jpeg-500x500.jpg"
}
```

```python
#Reading the image
image =cv2.imread(args["image"])
plt_imshow("Drawing",image)
```

```python
# Now getting the dimentions of the image
w=image.shape[1]
h=image.shape[0]
c=image.shape[2]
print("The height of the image is " + str(h))
print("The weight of the image is " + str(w))
print("The channel of the image is " + str(c))
```

```python
#cropping the image
(h, w) = image.shape[:2]
cx=h//2
cy=w//2
```

```python
tl = image[0:cy, 0:cx]
plt_imshow("Top-Left Corner", tl)
```

```python
#translate the image
shift=imutils.translate(image,-50,-20)
plt_imshow("Image",shift)
```

```python
(thresh, Image) = cv2.threshold(image, 127, 255, cv2.THRESH_BINARY)
```

```python
plt_imshow("greyimage",Image)
```

```python
plt.imshow(image,cmap="cool")
```

# **TXT file**(text.txt)

```python
!wget "https://raw.githubusercontent.com/ramahanisha-7/Datasets/main/text.txt"
```

```python
from wordcloud import WordCloud, STOPWORDS
text = open("text.txt", "r").read()
# Create and generate a word cloud image:

wordcloud = WordCloud(width = 3000, height = 2000, random_state=1, background_color='black', colormap='Pastel1', collocations=False, stopwords = STOPWORDS).generate(text)
# Display the generated image:
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
