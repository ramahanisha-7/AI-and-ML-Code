```python
import cv2
import numpy as np
import argparse
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image 
import glob, os
```

```python
def dividex4(img):
  height = img.shape[0]
  width = img.shape[1]
  # Cut the image in half
  width_cutoff = width // 2
  height_cutoff = height // 2

  left = img[:, :width_cutoff]
  right = img[:, width_cutoff:]
  # finish vertical devide image
  
  up_left = left[:height_cutoff, :]
  down_left = left[height_cutoff:, :]

  up_right = right[:height_cutoff, :]
  down_right = right[height_cutoff:, :]

  ls = [up_left,down_left,up_right,down_right]
  return ls
```

```python
def bi_histogram(img):
  #tm = path.split('/')
  equ = cv2.equalizeHist(img)
  #im = Image.fromarray(equ)
  #im.save('{}{}'.format(output, tm[-1]))
  return equ
```

```python
def stack(ls, path, output):
  tm = path.split('/')
  left = np.vstack([ls[0],ls[1]])
  right = np.vstack([ls[2],ls[3]])
  image = np.hstack([left, right])
  im = Image.fromarray(image)
  im.save('{}{}'.format(output, tm[-1]))
```
