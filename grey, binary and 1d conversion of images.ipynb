# Binary Image dataset (gray scale)

Here i converted the original image to grayscale and applied binary

```python
Binary_image=list()
```

```python
for i in pathsZombie:
  img=cv2.imread(i)
  img=cv2.resize(img,(100,100))
  image=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
  ret,bi_img=cv2.threshold(image,127, 255,cv2.THRESH_BINARY)
  Binary_image.append(bi_img)

```

```python
for j in pathsAlien:
  img=cv2.imread(j)
  img=cv2.resize(img,(100,100))
  image=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
  ret,bi_img=cv2.threshold(image,127, 255,cv2.THRESH_BINARY)
  Binary_image.append(bi_img)

```

# 1D vector dataset

```python
from PIL import Image
```

```python
vector_image=list()
```

```python
for i in pathsZombie:
  img = Image.open(i).convert('RGBA')
  arr = np.array(img)
  shape = arr.shape
  flat_arr = arr.ravel()
  vector = np.matrix(flat_arr)
  vector[:,::10] = 128
  arr2 = np.asarray(vector).reshape(shape)
  img2 = Image.fromarray(arr2, 'RGBA')
  img2.show()
  vector_image.append(img2)
```

```python
for j in pathsAlien:
  img = Image.open(j).convert('RGBA')
  arr = np.array(img)
  shape = arr.shape
  flat_arr = arr.ravel()
  vector = np.matrix(flat_arr)
  vector[:,::10] = 128
  arr2 = np.asarray(vector).reshape(shape)
  img2 = Image.fromarray(arr2, 'RGBA')
  img2.show()
  vector_image.append(img2)
```
